```python
# Experiment to change column names
employees = employees.rename(columns={'UserID': 'EmployeeID', 'First Name': 'FirstName', 'Join_Date': 'HireDate', 'Salary': 'Compensation', 'Department': 'Dept'})
employees.columns
```

```python
# change data type of EmployeeID to string

employees['User_ID'] = employees['User_ID'].astype(str)
```

```python
# check the data type again to confirm the change

employees.dtypes
```

```python
# check the dtypes again to confirm the change

employees.dtypes
```

```python
# check the stats

employees.describe()
```

```python
# testing to_numeric() function to convert Compensation to numeric

employees['Compensation'] = pd.to_numeric(employees['Compensation'], errors='coerce')
```

```python
# view data types again to confirm the change
employees.dtypes
```

```python
employees.describe()
```

```python
# convert blanks to null/na values
employees.replace('', pd.NA, inplace=True)
```

```python
# count nulls

employees.isnull().sum()
```

```python
# view nulls in user id column

employees[employees['User_ID'].isnull()]
```

```python
# employee with null user id has all other data so we keep the row and try to find the user ID from stakeholders.
```

```python
# view nulls in Compensation column

employees[employees['Compensation'].isnull()]
```

```python
# critical compensation data is missing so delete the row (if the compensation data cannot be found from stakeholders). delete the entire row for now.

employees.dropna(subset=['Compensation'], inplace=True)
```

```python
#  Recheck nulls in Compensation column to confirm the change

employees[employees['Compensation'].isnull()]
```

```python
# test for duplicate rows   

employees[employees.duplicated(keep=False)]
```

```python
# drop duplicate rows if any

employees.drop_duplicates(inplace=True)
```
